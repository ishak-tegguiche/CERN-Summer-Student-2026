# Introduction to Mechanics

## What is Mechanics? 
Mechanics is the branch of physics that studies the motion of objects and the forces that cause this motion.

The foundations of classical mechanics were established by Isaac Newton who formulated the laws of motion and universal gravitation. Later developments by Joseph-Louis Lagrange and William Rowan Hamilton reformulated classical mechanics into a powerful mathematical theory known as analytical mechanics.

A **particle** is an idealized object whose size and internal structure are neglected. It is treated as a point-like object with a position in space. In classical mechanics, its physical state is completely described by its position and velocity (or momentum), while its internal structure, orientation, and deformation are ignored.

## Kinematics and Dynamics 
Mechanics can be devided into two main branches 
1. **Kinematics** which describes motion without considering the forces that cause the motion. 
It focuses on quantities such as : position, velocity and acceleration and study how this quantities change with time
2. **Dynamics** which studies relationship between motion and the forces acting on a system. 
Newton's laws of motion provide the fundamental principles govering dynamics

### The State of a System 
To describe a physical system, we need to know its state. In classical mechanics, the state of a particle is completely described by:
1. **Position**

The position of a particle describes its location in space:

$$
\vec{x}(t)
$$

where $(\vec{x})$ is the position vector and $(t)$ is time.

2. **Velocity**

The velocity describes how the position changes with time:

$$
\vec{v}(t)=\frac{d\vec{x}(t)}{dt}
$$

where $(\vec{v})$ is the velocity vector

## Work and Energy
In many mechanical problems it is convenient to analyze motion using energy methods rather than directly applying Newton's laws  
Energy methods provide powerful tools for solving problems involving complicated forces or trajectories   

1. **Work** is the transfer of energy when a force causes an object to move.

Two conditions must be satisfied:

- A force must act on the object.
- The object must undergo a displacement.

For a constant force,

$$
W=\vec{F}\cdot\vec{d}=Fd\cos\theta,
$$

where:
- $(W)$ is the work,
- $(\vec{F})$ is the applied force,
- $(\vec{d})$ is the displacement,
- $(\theta)$ is the angle between the force and the displacement.

The work depends on : the magnitude of the force - the displacement - the angle between the force and the direction of motion 

- Variable Forces

When the force changes along the path, the work is given by

$$
W=\int \vec{F}\cdot d\vec{r}.
$$

where:
- $d\vec{r}$ is the displacement

2. **Energy** is the capacity of a physical system to perform work or produce change.  
It is a scalar quantity measured in **joules (J)**.

Two important forms of mechanical energy are:

- **Kinetic Energy**

Kinetic energy is the energy associated with motion.

For a particle of mass $(m)$ moving with velocity $(v)$,

$$
T=\frac{1}{2}mv^2.
$$

Kinetic energy increases with both mass and velocity.

- **Potential Energy**

Potential energy is the energy an object stores due to its position or configuration.

Examples include:

1. Gravitational Potential Energy

$$
V=mgh,
$$

where:

- $(m)$ is the mass,
- $(g)$ is the gravitational acceleration,
- $(h)$ is the height.

2. Elastic Potential Energy

For a spring,

$$
V=\frac{1}{2}kx^2,
$$

where:

- $(k)$ is the spring constant,
- $(x)$ is the displacement from equilibrium.

3. Electric Potential Energy

For two electric charges,

$$
V=\frac{kq_1q_2}{r}.
$$

where:

- $(k)$ is **Coulomb's constant**:$
k = 8.99 \times 10^9 \ \text{N}\cdot\text{m}^2/\text{C}^2$
- $(q_1$ is the charge of the first particle measured in coulombs (C).
- $(q_2)$ is the charge of the second particle measured in coulombs (C).
- $(r)$ is the distance between the two charges measured in meters (m).


- **Mechanical Energy**

The total mechanical energy is the sum of kinetic and potential energies:

$$
E=T+V.$$


---

##### Conservation of Energy

One of the most important principles in physics is the conservation of energy.

> Energy cannot be created or destroyed. It can only be transformed from one form to another.

For an isolated system,

$$
E_{\text{initial}} = E_{\text{final}}
$$

---

##### The Work-Energy Theorem

The work-energy theorem states that the net work done on an object equals the change in its kinetic energy:

$$
W_{\text{net}}=\Delta T.
$$

or

$$
W_{\text{net}}=T_f-T_i.
$$

This theorem provides a direct connection between force and motion through energy.

---

##### Why Energy is Important

Many physical systems can be described more easily using energy rather than forces.

This energy-based viewpoint leads naturally to:

- **Lagrangian Mechanics**, based on the difference between kinetic and potential energy,

$$
L=T-V,
$$

- **Hamiltonian Mechanics**, based on the total energy of the system,

$$
H=T+V.
$$

These formulations form the foundation of modern theoretical physics and quantum mechanics.

## Newtonian Mechanics

1. **Introduction to Dynamics**

In mechanics, **kinematics** describes how objects move without considering the causes of motion.  
**Dynamics** studies the relationship between motion and the forces that produce it.

The foundation of classical dynamics is given by **Newton's laws of motion**, which describe how forces affect the state of a physical system.

A force can change:

- The velocity of an object.
- The direction of motion.
- The energy of a system.

---

2. **Newton's First Law: Law of Inertia**

Newton's first law states:

> An object remains at rest or continues moving with constant velocity unless acted upon by an external force.

Mathematically:

$$
\vec{F}_{net}=0
$$

implies:

$$
\vec{v}=constant
$$

This means that without a net force, an object's motion does not change.

**Inertia** is the tendency of an object to resist changes in its state of motion.

The amount of inertia is related to the mass of an object:

- Objects with larger mass require more force to change their motion.
- Objects with smaller mass are easier to accelerate.

---

3. **Newton's Second Law: Relation Between Force and Motion**

Newton's second law provides the fundamental equation of dynamics:

$$
\vec{F}=m\vec{a}
$$

where:

- $(\vec{F})$ is the net force acting on the object.
- $(m)$ is the mass of the object.
- $(\vec{a})$ is the acceleration produced by the force.

This equation tells us that:

- A larger force produces a larger acceleration.
- A larger mass requires more force to obtain the same acceleration.

The acceleration is the rate of change of velocity:

$$
\vec{a}=\frac{d\vec{v}}{dt}
$$

Therefore:

$$
\vec{F}=m\frac{d\vec{v}}{dt}
$$

---

4. **Force and Momentum**

A more general form of Newton's second law is expressed using momentum.

Momentum is defined as:

$$
\vec{p}=m\vec{v}
$$

where:

- $(\vec{p})$ is the momentum.
- $(m)$ is the mass.
- $(\vec{v})$ is the velocity.

Newton's second law can then be written as:

$$
\vec{F}=\frac{d\vec{p}}{dt}
$$

This form is more general because it also applies when the mass is not constant.

The physical meaning is:

> A force changes the momentum of an object.

---

5. **Newton's Third Law: Action and Reaction**

Newton's third law states:

> When two objects interact, they exert forces on each other that are equal in magnitude and opposite in direction.

Mathematically:

$$
\vec{F}_{12}=-\vec{F}_{21}
$$

where:

- $(\vec{F}_{12})$ is the force exerted by object 1 on object 2.
- $(\vec{F}_{21})$ is the force exerted by object 2 on object 1.

---

6. **Connection Between Force and Energy**

Force and energy are closely related.

The work done by a force changes the kinetic energy of an object:

$$
W_{net}=\Delta T
$$

where:

$$
T=\frac{1}{2}mv^2
$$

is the kinetic energy.

Therefore, forces affect motion by transferring energy into or out of a system.

---

7. **Conservation of Momentum**

For an isolated system where no external forces act:

$$
\vec{F}_{ext}=0
$$

the total momentum remains constant:

$$
\vec{p}_{initial}=\vec{p}_{final}
$$

This principle explains many physical phenomena:

- Collisions between objects.
- Rocket propulsion.
- Interactions between particles.

---

8. **Conservation of Angular Momentum**

For rotational systems, another conserved quantity is angular momentum.

Angular momentum is defined as:

$$
\vec{L}=\vec{r}\times\vec{p}
$$

where:

- $(\vec{r})$ is the position vector.
- $(\vec{p})$ is the momentum.

If no external torque acts on the system:

$$
\frac{d\vec{L}}{dt}=0
$$

then:

$$\vec{L}=constant
$$


---

9. **Limitations of Newtonian Mechanics**

Newtonian mechanics successfully describes many macroscopic systems, but it has limitations:

- **High velocities**

When velocities approach the speed of light, Newtonian mechanics is replaced by **relativity**.

- **Microscopic systems**

At atomic and subatomic scales, particles do not follow classical trajectories. Their behavior is described by **quantum mechanics**.

- **Complex systems**

For systems containing many interacting particles, solving Newton's equations becomes extremely difficult.

These limitations motivated the development of more general formulations:

$$
\boxed{\text{Newtonian Mechanics}}
$$

$$
\downarrow
$$

$$
\boxed{\text{Lagrangian Mechanics}}
$$

$$
\downarrow
$$

$$
\boxed{\text{Hamiltonian Mechanics}}
$$

which provide the foundation for modern physics.

## Lagrangian Mechanics

1. **Why Do We Need a New Formulation?**

Newtonian mechanics successfully describes the motion of many physical systems using forces. The motion of a particle is determined by Newton's second law,

$$
\vec{F}=m\vec{a}.
$$

Although this approach is simple for a single particle, it becomes increasingly difficult for systems involving many interacting particles or constraints.

Describing all these systems directly with forces often requires introducing constraint forces that are difficult to determine.

Physicists therefore developed a more general formulation based on **energy** rather than forces. This formulation is known as **Lagrangian mechanics** and provides a unified framework that applies to classical mechanics, relativity, and quantum field theory.

2. **Generalized Coordinates**

In Newtonian mechanics, the position of a particle is usually described using **Cartesian coordinates**:

$$
(x, y, z).
$$

These coordinates work well for simple systems moving freely in space. However, many physical systems are subject to **constraints**, meaning their motion is restricted.

Using Cartesian coordinates for these systems introduces unnecessary variables and additional constraint equations.

To simplify the description, Lagrangian mechanics introduces **generalized coordinates**, denoted by

$$
q_1, q_2, \ldots, q_n.
$$

A generalized coordinate is **any variable that uniquely describes the configuration of a system while automatically satisfying its constraints**.

Unlike Cartesian coordinates, generalized coordinates do not have to represent distances. They can be:

- Linear positions
- Angles
- Lengths
- Any other variables that completely describe the system

The number of generalized coordinates equals the number of **degrees of freedom** of the system.  
The **degrees of freedom (DOF)** of a system are the **minimum number of independent variables required to completely describe its configuration or state**.  
In other words, they represent the number of independent ways in which a system can move.

- **Example: Free Particle**

A particle moving freely in three-dimensional space requires three coordinates:

$$
q_1=x,\qquad q_2=y,\qquad q_3=z.
$$

Thus, the particle has **three degrees of freedom**.


- **Generalized Velocities**

Each generalized coordinate has a corresponding generalized velocity, defined as

$$
\dot{q}_i=\frac{dq_i}{dt}.
$$

The complete state of the system in Lagrangian mechanics is described by

$$
(q_1,q_2,\ldots,q_n,\dot{q}_1,\dot{q}_2,\ldots,\dot{q}_n).
$$

These variables replace the Cartesian coordinates used in Newtonian mechanics and provide a more natural description of constrained systems.

- **Why Generalized Coordinates?**

The use of generalized coordinates offers several advantages:

- They automatically satisfy the constraints of the system.
- They reduce the number of variables required to describe the motion.
- They simplify the mathematical equations of motion.
- They provide a framework that extends naturally to advanced theories such as relativity and quantum field theory.

Generalized coordinates form the foundation of Lagrangian mechanics and are the variables used to construct the **Lagrangian** of a physical system.


3. **Principle of Least Action**

- **Why Do We Need Another Principle?**

Newtonian mechanics describes the motion of a system by analyzing the **forces** acting on it. While this approach is very successful, it is not always the most convenient, especially for systems with many particles or constraints.

Lagrangian mechanics takes a different approach. Instead of focusing on forces at every instant of time, it considers the **entire motion** of the system between two points in time.

The central question becomes:

> **Among all the possible paths connecting two points, which one does nature choose?**

The answer is given by the **Principle of Least Action**.

---

- **The Idea of Action**

Imagine throwing a ball from one point to another.

There are infinitely many paths that could connect the starting point and the final point.

However, in nature, the ball follows only **one physical path**.

The quantity used to determine this path is called the **action**, denoted by $(S)$.

The action is defined as

$$
S=\int_{t_1}^{t_2}L\,dt,
$$

where:

- $(S$) is the action,
- $(L)$ is the Lagrangian,
- $(t_1)$ and $(t_2)$ are the initial and final times.

The action measures the total dynamical behavior of the system along an entire trajectory.

---

- **The Principle of Least Action**

The Principle of Least Action states:

> **The physical path followed by a system is the one for which the action is stationary (usually a minimum) compared with all nearby possible paths.**

Mathematically,

$$
\delta S=0.
$$

This means that if we slightly change the path while keeping the starting and ending points fixed, the action does not change to first order.

Although the name suggests that the action is always minimized, this is not strictly true. In general, the action is **stationary**, meaning it can correspond to a minimum, a maximum, or another stationary value.

---

- **Physical Interpretation**

The principle does **not** mean that nature "tries" to minimize anything.

Instead, it is a mathematical principle that correctly predicts the motion of physical systems.

Among all imaginable trajectories, only one satisfies the condition

$$
\delta S=0,
$$

and this trajectory is the one observed experimentally.

---

- **Why Is the Principle Important?**

The Principle of Least Action is one of the most fundamental ideas in theoretical physics.

It provides a unified way to derive the equations of motion for a wide variety of physical systems.

Many important theories are based on this principle, including:

- Classical mechanics
- Electromagnetism
- General relativity
- Quantum mechanics
- Quantum field theory

Rather than introducing different equations for different systems, the Principle of Least Action provides a single mathematical framework from which these equations can be derived.

---

- **From Action to the Lagrangian**

The action depends on a quantity called the **Lagrangian**.

4. **The Lagrangian**

- **Definition of the Lagrangia**

The Lagrangian is a function that describes the dynamics of a physical system.

It is generally written as:

$$
L(q_i,\dot{q}_i,t)
$$

where:

- $(q_i)$ are the generalized coordinates.
- $(\dot{q}_i)$ are the generalized velocities.
- $(t)$ is time.

For many classical systems, the Lagrangian is defined as:

$$
\boxed{L=T-V}
$$

where:

- $(T)$ is the kinetic energy of the system.
- $(V)$ is the potential energy of the system.

---

- **Physical Meaning of the Lagrangian**

The Lagrangian is not the energy of the system.

Instead, it represents the difference between two forms of energy:

$$
\text{Lagrangian}=\text{Kinetic energy}-\text{Potential energy}
$$

The kinetic energy describes the motion of the system, while the potential energy describes interactions or constraints.

The system evolves in such a way that the action:

$$S=\int L\,dt$$


becomes stationary.

---

- Example: Free Particle

Consider a particle of mass $(m)$ moving freely without any potential energy.

The kinetic energy is:

$$
T=\frac12m\dot{x}^2
$$

Since:

$$
V=0
$$

the Lagrangian is:

$$
L=T-V
$$

therefore:

$$
\boxed{L=\frac12m\dot{x}^2}
$$

The Lagrangian contains all the information needed to determine the particle's motion.

---

- Example: Particle in a Potential Field

Now consider a particle moving under a potential $(V(x))$.

The kinetic energy is:

$$
T=\frac12m\dot{x}^2
$$

The Lagrangian becomes:

$$
\boxed{
L=\frac12m\dot{x}^2-V(x)
}
$$

The first term describes the motion, and the second term describes the interaction with the environment.

---

- **Why Use \(L=T-V\)?**

The choice:

$$
L=T-V
$$

may look arbitrary at first, but it is precisely the quantity that produces the correct equations of motion.

When the condition:

$$
\delta S=0
$$

is applied to the action:

$$
S=\int Ldt,
$$

it leads to the Euler–Lagrange equation:

$$
\frac{d}{dt}
\left(
\frac{\partial L}{\partial \dot q_i}
\right)
-\frac{\partial L}{\partial q_i}
=0.
$$

This equation reproduces Newton's laws when applied to classical systems.

---

- **Lagrangian vs Newtonian Description**

| Newtonian Mechanics | Lagrangian Mechanics |
|---|---|
| Based on forces | Based on energy |
| Uses \(F=ma\) | Uses \(L=T-V\) |
| Describes forces acting on particles | Describes the whole system |
| Uses Cartesian coordinates | Uses generalized coordinates |
| Difficult with constraints | Naturally handles constraints |

---


5. **Euler–Lagrange Equations**

- **From Action to Equations of Motion**

The Principle of Stationary Action states that the physical path of a system is the one that makes the action stationary:

$$
\delta S=0
$$

where the action is defined as:

$$
S=\int_{t_1}^{t_2}L(q_i,\dot{q}_i,t)\,dt.
$$

By applying the condition of stationary action, we obtain the **Euler–Lagrange equation**:

$$
\boxed{
\frac{d}{dt}
\left(
\frac{\partial L}{\partial \dot{q}_i}
\right)
-\frac{\partial L}{\partial q_i}
=0
}
$$

This equation determines the evolution of the generalized coordinates $(q_i)$.

---

- **Meaning of Each Term**

The Euler–Lagrange equation contains two important derivatives.

- Derivative with respect to position

$$
\frac{\partial L}{\partial q_i}
$$

This describes how the Lagrangian changes when the position (generalized coordinate) changes.

It is related to forces acting on the system.

---

- Derivative with respect to velocity

$$
\frac{\partial L}{\partial \dot{q}_i}
$$

This quantity is related to the momentum of the system.

It is called the **generalized momentum**:

$$
p_i=\frac{\partial L}{\partial \dot{q}_i}.
$$

---

- **Physical Interpretation**

The Euler–Lagrange equation states that:

> The change of generalized momentum with time is determined by how the Lagrangian depends on the generalized coordinates.

It plays the same role in Lagrangian mechanics as Newton's second law:

$$
F=ma
$$

plays in Newtonian mechanics.

---

- **Recovering Newton's Second Law**

Consider a particle moving in one dimension.

The Lagrangian is:

$$
L=T-V
$$

with:

$$
T=\frac12m\dot{x}^2
$$

and

$$
V=V(x).
$$

Therefore:

$$
L=\frac12m\dot{x}^2-V(x)
$$

---

- **Step 1: Calculate the velocity derivative**

$$
\frac{\partial L}{\partial \dot{x}}
=m\dot{x}
$$

Taking the time derivative:

$$
\frac{d}{dt}
\left(
\frac{\partial L}{\partial \dot{x}}
\right)=
m\ddot{x}
$$

---

- **Step 2: Calculate the position derivative**

$$
\frac{\partial L}{\partial x}
=-\frac{\partial V}{\partial x}
$$

The force is related to the potential by:

$$
F=-\frac{\partial V}{\partial x}
$$

Therefore:

$$
\frac{\partial L}{\partial x}=F
$$

---

- **Step 3: Apply Euler–Lagrange Equation**

$$
\frac{d}{dt}
\left(
\frac{\partial L}{\partial \dot{x}}
\right)
-\frac{\partial L}{\partial x}
=0
$$

Substituting:

$$
m\ddot{x}-F=0
$$

which gives:

$$
\boxed{F=m\ddot{x}}
$$

We recover Newton's second law.

---

- **Why Is This Important?**

The Euler–Lagrange equation shows that:

- Newtonian mechanics and Lagrangian mechanics describe the same physics.
- Forces do not need to be explicitly calculated.
- The equations of motion come from the structure of the Lagrangian.

This is why Lagrangian mechanics is more powerful for complex systems.

---

- Example: Simple Pendulum

For a pendulum, the generalized coordinate is the angle:

$$
q=\theta
$$

The kinetic energy is:

$$
T=\frac12ml^2\dot{\theta}^2
$$

The potential energy is:

$$
V=mgl(1-\cos\theta)
$$

Therefore:

$$
L=
\frac12ml^2\dot{\theta}^2
-mgl(1-\cos\theta)
$$

Applying the Euler–Lagrange equation gives:

$$
ml^2\ddot{\theta}+mgl\sin\theta=0
$$

or:

$$
\ddot{\theta}+\frac{g}{l}\sin\theta=0
$$

which is the equation describing the motion of the pendulum.

---

- **Summary**

The Lagrangian approach follows this chain:

$$
\boxed{
L=T-V
}
$$

↓

$$
\boxed{
S=\int Ldt
}
$$

↓

$$
\boxed{
\delta S=0
}
$$

↓

$$
\boxed{
\frac{d}{dt}
\frac{\partial L}{\partial\dot q_i}
-\frac{\partial L}{\partial q_i}
=0
}
$$
↓
$$
\boxed{\text{Equations of motion}}
$$

The Euler–Lagrange equation is the fundamental equation of Lagrangian mechanics and the starting point for more advanced formulations of physics.

## Hamiltonian Mechanics

1. **Why Do We Need Another Formulation?**

Lagrangian mechanics provides a powerful description of motion using generalized coordinates and velocities. From the Lagrangian,

$$
L(q_i,\dot{q}_i,t),
$$

the equations of motion are obtained through the Euler–Lagrange equations.

Although this formulation is very general, many areas of modern physics especially quantum mechanics are more naturally described in terms of **coordinates** and **momenta** rather than coordinates and velocities.

For this reason, Hamiltonian mechanics reformulates the dynamics of a system using:

- Generalized coordinates:

$$
q_i
$$

- Generalized momenta:

$$
p_i
$$

instead of the generalized velocities $(\dot{q}_i)$.

The state of the system is therefore represented by the pair

$$
(q_i,p_i),
$$

which defines the **phase space** of the system.

Hamiltonian mechanics provides a mathematically elegant framework that is widely used in classical mechanics, statistical mechanics, quantum mechanics, and quantum field theory.

---

2. **Generalized Momentum**

Before defining the Hamiltonian, we first introduce the concept of **generalized momentum**.

For each generalized coordinate $(q_i)$, the corresponding generalized momentum is defined as

$$
\boxed{
p_i=\frac{\partial L}{\partial\dot{q}_i}
}
$$

Unlike the classical momentum

$$
\vec{p}=m\vec{v}
$$

the generalized momentum depends on the form of the Lagrangian.

For many mechanical systems,

$$
p_i=m\dot{q}_i
$$

but in more general systems such as charged particles in electromagnetic fields or relativistic systems, the generalized momentum contains additional terms.

Generalized momentum is therefore a more universal concept than classical momentum.

---

3. **From the Lagrangian to the Hamiltonian**

The Lagrangian depends on

$$
(q_i,\dot{q}_i,t).
$$

Hamiltonian mechanics replaces the velocities by their corresponding generalized momenta.

This change of variables is performed through a mathematical procedure called the **Legendre transformation**.

The Hamiltonian is defined as

$$
\boxed{
H(q_i,p_i,t)
=\sum_i p_i\dot{q}_i-L
}
$$

where

- $(p_i)$ is the generalized momentum,
- $(\dot{q}_i)$ is the generalized velocity,
- $(L)$ is the Lagrangian.

The Hamiltonian is therefore obtained directly from the Lagrangian.

---

4. **Physical Meaning of the Hamiltonian**

For many classical systems,

$$
\boxed{
H=T+V
}
$$

where

- $(T)$ is the kinetic energy,
- $(V)$ is the potential energy.

Thus, the Hamiltonian represents the **total energy** of the system.

It is important to note that this equality is valid for many conservative mechanical systems but is **not universal**. In more general systems, the Hamiltonian may not be equal to the total mechanical energy.

Nevertheless, in many applications including quantum mechanics the Hamiltonian represents the total energy and governs the time evolution of the system.

---

- Example

Consider a free particle of mass \(m\).

The Lagrangian is

$$
L=\frac12m\dot{x}^2.
$$

The generalized momentum is

$$
p=\frac{\partial L}{\partial\dot{x}}
=m\dot{x}.
$$

Therefore,

$$
\dot{x}=\frac{p}{m}.
$$

Substituting into the Hamiltonian,

$$
H=p\dot{x}-L,
$$

gives

$$
H
=p\left(\frac{p}{m}\right)
-\frac12m
\left(\frac{p}{m}\right)^2.
$$

Simplifying,

$$
\boxed{
H=\frac{p^2}{2m}
}
$$

which is simply the kinetic energy of a free particle.

This illustrates that the Hamiltonian often corresponds to the total energy of the system.

5. **Hamilton's Equations of Motion**

- **From the Hamiltonian to the Motion**

Once the Hamiltonian of a system is known, its time evolution is completely determined by **Hamilton's equations**.

These equations replace Newton's laws and the Euler–Lagrange equations by describing the evolution of the generalized coordinates and generalized momenta.

For a system with generalized coordinates $(q_i)$ and generalized momenta $(p_i)$, Hamilton's equations are

$$
\boxed{
\dot{q}_i=\frac{\partial H}{\partial p_i}
}
$$

and

$$
\boxed{
\dot{p}_i=-\frac{\partial H}{\partial q_i}
}
$$

Together, these two equations completely describe the dynamics of the system.

---

- **Physical Meaning**

Hamilton's equations describe how the state of a system changes with time.

The first equation,

$$
\dot{q}_i=\frac{\partial H}{\partial p_i}
$$

tells us how the **position (or generalized coordinate)** changes.

The second equation,

$$
\dot{p}_i=-\frac{\partial H}{\partial q_i}
$$

tells us how the **momentum** changes.

Thus, instead of solving Newton's equation

$$
F=ma
$$

we solve two first-order differential equations that describe the evolution of the coordinates and momenta.

---

- **Phase Space**

Unlike Newtonian mechanics, which describes motion in ordinary space, Hamiltonian mechanics describes motion in **phase space**.

A point in phase space is defined by

$$
(q_i,p_i)
$$

For every degree of freedom, there are:

- one generalized coordinate,
- one generalized momentum.

Therefore, a system with \(n\) degrees of freedom has a **2n-dimensional phase space**.

Examples:

- One-dimensional particle:

$$
(x,p)
$$

→ 2-dimensional phase space.

- Particle in three-dimensional space:

$$
(x,y,z,p_x,p_y,p_z)
$$

→ 6-dimensional phase space.

As time evolves, the system moves through phase space according to Hamilton's equations.

---

- Example: Free Particle

Consider a free particle with Hamiltonian

$$
H=\frac{p^2}{2m}.
$$

- **First Hamilton Equation**

$$
\dot{x}
=\frac{\partial H}{\partial p}
=\frac{p}{m}.$$


Since


$$p=mv$$

we obtain

$$
\dot{x}=v$$

which is the familiar expression for velocity.

---

- **Second Hamilton Equation**

Since

$$
H=\frac{p^2}{2m}
$$

does not depend on the position \(x\),

$$
\frac{\partial H}{\partial x}=0.
$$

Therefore,

$$
\dot{p}=0.
$$

This means that the momentum remains constant, exactly as expected for a free particle experiencing no external force.

---

- **Hamiltonian as the Generator of Time Evolution**

One of the most important ideas in physics is that the Hamiltonian generates the time evolution of a system.

Knowing the Hamiltonian allows us to determine how the coordinates and momenta evolve over time through Hamilton's equations.

This idea extends naturally to quantum mechanics, where the Hamiltonian operator determines the time evolution of the quantum state through the Schrödinger equation.

---

- **Comparison of the Three Formulations**

| Formulation | Main Quantity | Equation of Motion |
|-------------|---------------|--------------------|
| Newtonian Mechanics | Force $(\mathbf{F})$ | $(\mathbf{F}=m\mathbf{a})$ |
| Lagrangian Mechanics | Lagrangian $(L=T-V)$ | Euler–Lagrange equation |
| Hamiltonian Mechanics | Hamiltonian $(H)$ | Hamilton's equations |

Although these formulations use different mathematical approaches, they all describe the same physical motion.

---

- **Why Hamiltonian Mechanics Is Important**

Hamiltonian mechanics offers several advantages:

- It provides a symmetric description of coordinates and momenta.
- It naturally describes systems in phase space.
- It simplifies the study of complex dynamical systems.
- It forms the foundation of statistical mechanics.
- It is the starting point of quantum mechanics.

In modern physics, the Hamiltonian is one of the most fundamental quantities because it determines the evolution of both classical and quantum systems.

---

- **Transition to Quantum Mechanics**

In classical mechanics, the Hamiltonian is a function of the coordinates and momenta:

$$
H(q_i,p_i).
$$

In quantum mechanics, these quantities become **operators**, and the Hamiltonian becomes the **energy operator** of the system.

The evolution of a quantum state is no longer described by Hamilton's equations but by the **Schrödinger equation**:

$$
i\hbar\frac{\partial}{\partial t}|\psi(t)\rangle
=\hat{H}|\psi(t)\rangle.
$$

This equation is the quantum analogue of Hamilton's equations and marks the transition from classical to quantum mechanics.

## The Hamiltonian Spectrum

1. **From Classical to Quantum Mechanics**

In classical mechanics, the Hamiltonian is a function of the generalized coordinates and momenta,

$$
H(q_i,p_i),
$$

and it determines how the system evolves with time through Hamilton's equations.

In quantum mechanics, the generalized coordinates and momenta become **operators**, and the Hamiltonian becomes the **Hamiltonian operator**,

$
\hat{H},
$
which represents the total energy of the quantum system.

Unlike classical mechanics, where the energy can vary continuously, quantum mechanics predicts that many systems can only possess **specific allowed energy values**.

These allowed energies form the **energy spectrum** of the Hamiltonian.

---

2. **Eigenvalues and Eigenstates**

To determine the allowed energies, we solve the eigenvalue equation

$$
\boxed{
\hat{H}|\psi_n\rangle
=E_n|\psi_n\rangle
}
$$

where:

- $(\hat{H})$ is the Hamiltonian operator.
- $(|\psi_n\rangle)$ is an eigenstate (or energy eigenstate).
- $(E_n)$ is the corresponding eigenvalue (energy).

This equation states that when the Hamiltonian acts on one of its eigenstates, the state does not change. It is only multiplied by its corresponding energy.

Each eigenstate therefore represents a quantum state with a well-defined energy.

---

3. **The Energy Spectrum**

The collection of all eigenvalues

$$
E_0,E_1,E_2,\ldots
$$

is called the **energy spectrum** of the Hamiltonian.

Each energy level corresponds to one or more eigenstates.

The spectrum contains all the information about the possible energies of the quantum system.

---

4. **Ground State**

The eigenstate with the **lowest energy** is called the **ground state**.

Its energy is
$
E_0.
$

The ground state is the most stable state of the system.

At absolute zero temperature, a system naturally occupies its ground state.

---

4. **Excited States**

All eigenstates with energies larger than the ground-state energy are called **excited states**.

Their energies satisfy

$$
E_1>E_0,
$$

$$
E_2>E_1,
$$

and so on.

A system can move from one energy level to another by absorbing or releasing energy.

---

5. **Energy Gap**

An important quantity is the **energy gap**, defined as the difference between the ground-state energy and the first excited-state energy.

It is given by

$$\boxed{
\Delta E=E_1-E_0.
}$$

The energy gap plays an important role in many physical phenomena.

- A **large energy gap** means that exciting the system requires a significant amount of energy.
- A **small or zero energy gap** means that the system can be excited more easily.

The presence or absence of an energy gap is often used to distinguish different phases of matter.

---

6. **Why Do We Need the Spectrum?**

Knowing the Hamiltonian is not enough.

To understand the behavior of a quantum system, we must determine:

- The allowed energy levels.
- The corresponding eigenstates.

These quantities tell us:

- The lowest-energy configuration.
- How the system responds to external perturbations.
- How the system evolves with time.

The spectrum therefore provides a complete description of the energetic properties of the system.
